Para carregar um dataset no Numpy, utilizamos o método `genfromtxt()`.  
Nós podemos especificar o nome do arquivo, delimitador, número de linhas a serem puladas e dtype.

```python
import numpy as np
```

```python
wines=np.genfromtxt("../datasets/winequality-red.csv",delimiter=";",skip_header=1)
wines
```

### Agora vamos aplicar algumas coisas que foram estudadas

Podemos selecionar alguma coluna ou linha especificamente.  
Se quisermos, por exemplo, a coluna 0 que representa o "fixed acidity".

```python
wines[:,0]
```

Podemos calcular a média da qualidade dos vinhos, selecionando a coluna correspondente e aplicando o método `mean`.

```python
wines[:,-1].mean()
```

## Data set sobre graduação

Dataset sobre graduação.  
Os campos incluem:  
`GRE score`  
`TOEFL scorte`  
`university rating`  
`GPA`  
`having research experience`  
`chance of admission`

Nós podemos especificar os nomes das colunas utilizando o parâmetro "names".

```python
graduate_admission=np.genfromtxt('../datasets/Admission_Predict.csv'
                                 ,delimiter=','
                                 ,skip_header=1
                                ,names=('Serial No','GRE Score', 'TOEFL Score', 'University Rating', 'SOP',
                                          'LOR','CGPA','Research', 'Chance of Admit'))
graduate_admission
```

Podemos percerbe que o resultado é um array unidimensional com 400 elementos e, portanto, utilizar [:,0] para pegar uma coluna não funciona.

Podemos pegar uma coluna utilizando o nome dela.

```python
graduate_admission['CGPA']
```

Podemos fazer um tratamento de dados na coluna GPA.  
Como o GPA vai de 1 a 10, e nos EUA é mais comum escalar até 4, podemos converter os valores.

```python
graduate_admission['CGPA'] = graduate_admission['CGPA'] /10 *4
graduate_admission['CGPA'][:5]
```

Podemos utilizar uma máscara boolean para descobrir quantos estudantes tem experiência em pesquisa.

```python
len(graduate_admission[graduate_admission['Research']==1])
```

Podemos calcular a média do CGPA e do GRE score para os alunos que tem alta chance de admissão (>0.8) e baixa chance de admissão (<0.4).

```python
print(graduate_admission[graduate_admission['Chance_of_Admit']>0.8]['CGPA'].mean())
print(graduate_admission[graduate_admission['Chance_of_Admit']<0.4]['CGPA'].mean())
```

```python
print(graduate_admission[graduate_admission['Chance_of_Admit']>0.8]['GRE_Score'].mean())
print(graduate_admission[graduate_admission['Chance_of_Admit']<0.4]['GRE_Score'].mean())
```
