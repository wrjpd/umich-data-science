Vamos limpar um data set

```python
import pandas as pd
df=pd.read_csv('datasets/presidents.csv')
```

```python
df.head()
```

```python
#Vamos começar limpando o nome do presidente e separando em primeiro e ultimo nome
# Então queremos criar 2 novas colunas e aplicarum regex que projete a coluna 'Presidente'

#Poderiamos fazer
df["First"]=df["President"]

# E então utilizar a função replace() para fazer o tratamento
df["First"]=df["First"].replace("[ ].*","",regex=True)
df.head()
```

```python
del  df["First"]
```

```python
# Esse jeito funciona mas é um jeito ruim e devagar.
# Vamos ver outro jeito utilizando a função apply()

# A função apply() recebe uma função arbitrária que criamos e aplica isso a uma Series ou a um Dataframe sobre todas as linhas ou colunas.
# Vamos criar uma função que splita o nome

def splitname(row):    
    # A linha é uma Series
    # Extraindo o primeiro nome
    row['First']=row['President'].split(" ")[0]
    # Vamos extrair o segundo nome
    row['Last']=row['President'].split(" ")[-1]
    # Now we just return the row and the pandas .apply() will take of merging them back into a DataFrame
    return row

# Agora utilizamos o apply indicando que queremos adicionar uma coluna:
df=df.apply(splitname, axis='columns')
df.head()
```

```python
# Outra função que podemos utilizar é a extract
# Vamos deletar as colunas criadas
del df["First"]
del df["Last"]
```

```python
df
```

```python
# A função extract usa uma expressão regular como entrada e requer especificamente que você defina grupos de captura que
# correspondem às colunas de saída nas quais você está interessado.

"?: significa 'grupo não capturado'"

pattern="(^[\w]*)(?:.* )([\w]*$)"

# A função extract() é criado dentro do atributo str de uma Series Object,então:
df['President'].str.extract(pattern).head()
```

```python
# Mas queremos nomear as colunas, então nomeamos os grupos
pattern="(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

# E chamamos a função extract
names=df["President"].str.extract(pattern).head()
names
```

```python
# E então adicionamos ao dataframe original
df["First"]=names["First"]
df["Last"]=names["Last"]
df.head()
```

```python
# Existem outras funções para o módulo "str" que foram escritos especificamente para limpar strings no DataFrame.
# https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html
```

```python
# Agora vamos limpar a coluna 'Born'. Vamos remover tudo que está fora do padrão de data:
df["Born"]=df["Born"].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
df["Born"].head()
```

```python
# Temos dtype=object
# Pandas tem uma feature muito interessante
df["Born"]=pd.to_datetime(df["Born"])
df["Born"].head()
```
